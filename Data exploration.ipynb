```python
# Preamble 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as dates
```

```python
# Load data
shark = pd.read_csv('18224_all.csv')
shark.head(5)
```

```python
# Add new columns for rounded lat / lon
shark['lat_round'] = shark.Latitude.round(decimals=2)
shark['lon_round'] = shark.Longitude.round(decimals=2)
shark.head()
```

```python
# Change index to datetime 
shark.set_index('datetime_utc', inplace=True)
shark.tail()
```

```python
# check how many unique lat/lons

#pd.unique(shark.lat_round)
pd.unique(shark.lon_round)
```

```python
# Make a plot!
import matplotlib.dates as mdates

fig, ax = plt.subplots()
fig.set_size_inches(16,6)
 
ax.plot(shark.lat_round, label=shark.lon_round)

#days = dates.DayLocator()
#months = dates.MonthLocator()
#daysFmt = dates.DateFormatter('%d')
#monthsFmt = dates.DateFormatter('%b')
#ax.xaxis.set_major_locator(months)

xfmt = mdates.DateFormatter('%M-%d')
ax.xaxis.set_major_formatter(xfmt)

#ax.xaxis.set_minor_locator(days)
#ax.xaxis.set_minor_formatter(daysFmt)

#plt.tight_layout()

#plt.setp(ax.xaxis.get_majorticklabels(), rotation=90)
```
